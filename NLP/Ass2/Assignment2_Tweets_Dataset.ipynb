# Assignment 2: NLTK vs spaCy Text Preprocessing

Dataset: Tweets.csv

```

import pandas as pd
df = pd.read_csv('Tweets.csv')
print(df.shape)
df.head()
```

```

TEXT_COLUMN = 'text'
df[TEXT_COLUMN].head()
```

```

import re, nltk, spacy, time, json, numpy as np, matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'https?://\S+', ' ', text)
    text = re.sub(r'@\w+', ' ', text)
    text = re.sub(r'#(\w+)', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```

sample = clean_text(df[TEXT_COLUMN].iloc[0])
print(sample)
```

```

stop_words=set(stopwords.words('english'))-{'not','no','never'}
lemmatizer=WordNetLemmatizer()

def nltk_preprocess(text):
    tokens=word_tokenize(clean_text(text))
    tokens=[t for t in tokens if t not in stop_words]
    lemmas=[lemmatizer.lemmatize(t) for t in tokens]
    return lemmas
```

```

# !python -m spacy download en_core_web_sm
import spacy
nlp=spacy.load('en_core_web_sm')

def spacy_preprocess(text):
    doc=nlp(clean_text(text))
    return [t.lemma_ for t in doc if not t.is_stop and not t.is_punct]
```

```

df['clean_text']=df[TEXT_COLUMN].astype(str).apply(clean_text)
df[['clean_text']].head()
```

```

lengths=df['clean_text'].str.len()
plt.hist(lengths,bins=30)
plt.title('Tweet Length Distribution')
plt.show()
```
